```python
import pickle
import json

import os

def is_json(myjson):
  try:
    json_object = json.loads(myjson)
  except ValueError as e:
    return False
  return True

def convert(model, type):
  path = f"data/fact_score_cleaned/42/output/{model}_{type}.pkl"

  input_path = f"data/fact_score_cleaned/42/input/{type}.pkl"


  with open(path, 'rb') as f:
      data = pickle.load(f)
      
      
  with open(input_path, 'rb') as f:
      input_data = pickle.load(f)

  data = [d[1].outputs[0].text for d in data]


  _, _ , topics = zip(*input_data)

  print(data[1])
  print(len(data))
  if type == "structured":
    is_structured = True
  else:
    is_structured = False


  if is_structured:
      # output = [json.loads(d)['answer'] for d in data if not is_json(d)]
      
      zipped = [(t, json.loads(d)) for t,d in zip(topics, data) if is_json(d)]
      zipped = [(t, d["bio"]) for t,d in zipped if 'bio' in d.keys()]
      print(len(zipped))
      # zipped = [(t, ) for t,d in zip(topics, data) ]
      # _, _ , topic = zip(*input_data)
  else:
    zipped = zip(topics, data)


  os.makedirs("data/fact_score_cleaned_converted", exist_ok=True)

  with open(f"data/fact_score_cleaned_converted/{model}_{type}.jsonl", 'w') as f:
      for t, o in zipped:
          f.write(json.dumps({'topic': t, 'output': o}) + '\n')


for model in ("phi3", "mistral", "mistral-instruct", "olmo", "olmo-instruct", "llama31-instruct", "llama31-base", "deepseek", "deepseek-chat", "deepseek-coder", "deepseek-coder-instruct"):
  for type in ("structured", "unstructured"):
    convert(model, type)
```
